# Image Super Resolution Pathfinder

## Dependencies

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = [7, 7]
```

## The Image being used:

```python
img = cv2.imread("bird.jpg")
plt.title('Initial Input Image - A Cockatiel(256x256px)')
plt.imshow(img[:,:,::-1])
```

```python
img1 = img[2:115,80:210]
plt.imshow(img1[:,:,::-1])
plt.title('Cropping out the cockatiel\'s head')
plt.show()
```

# Comparing Different Algorithms: 

## 1. EDSR

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
path = "models\EDSR\EDSR_x4.pb"
sr.readModel(path)
sr.setModel("edsr",4)
result = sr.upsample(img1)
resized = cv2.resize(img1,dsize=None,fx=3,fy=3)

fig, ax = plt.subplots(1, 3, figsize=(15, 15))
ax[0].imshow(img1[:,:,::-1])
ax[0].set_title('Original Image')

ax[1].imshow(result[:,:,::-1])
ax[1].set_title('Upscaled Image(EDSRx4)')

ax[2].imshow(resized[:,:,::-1])
ax[2].set_title('Resized Image using OpenCV')
```

## 2. FSRCNN

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
path = "models/FSRCNN/FSRCNN_x3.pb"
sr.readModel(path)
sr.setModel("fsrcnn",3)
result = sr.upsample(img1) # Upsampled Image
resized = cv2.resize(img1,dsize=None,fx=3,fy=3) # Resized Image

fig, ax = plt.subplots(1, 3, figsize=(15, 15))
ax[0].imshow(img1[:,:,::-1])
ax[0].set_title('Original Image')

ax[1].imshow(result[:,:,::-1])
ax[1].set_title('Upscaled Image(FSRCNNx3)')

ax[2].imshow(resized[:,:,::-1])
ax[2].set_title('Resized Image using OpenCV')
```

## 3. LapSRN

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()
path = "models/LapSRN/LapSRN_x8.pb"
sr.readModel(path)
sr.setModel("lapsrn",8)
result = sr.upsample(img1) # Upsampled Image
resized = cv2.resize(img1,dsize=None,fx=8,fy=8) # Resized image

fig, ax = plt.subplots(1, 3, figsize=(15, 15))
ax[0].imshow(img1[:,:,::-1])
ax[0].set_title('Original Image')

ax[1].imshow(result[:,:,::-1])
ax[1].set_title('Upscaled Image(LapSRNx8)')

ax[2].imshow(resized[:,:,::-1])
ax[2].set_title('Resized Image using OpenCV')
```
